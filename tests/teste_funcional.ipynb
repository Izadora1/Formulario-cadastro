```python
import pyautogui, time, pandas as pd
arquivo = R'D:\Izadora\ADS\3 período - ADS\Teste-manutencao\Form-cadastro\invalidos-para-preencher.xlsx'
df = pd.read_excel(arquivo)
#df['Data de Nascimento'] = df['Data de Nascimento'].dt.strftime('%d/%m/%Y')

def wait():
    time.sleep(0.3)

def repetir():
    wait()
    pyautogui.click(200,200)
    pyautogui.press("tab", presses=2)
    wait()
    
def aguardar():
    pyautogui.press("tab")
    wait()

pyautogui.hotkey("ctrl","t")
wait()
pyautogui.write("http://127.0.0.1:5500")
wait()
pyautogui.press("enter")
wait()

for _, row in df.iterrows():
    repetir()
    pyautogui.write(row['Nome Completo'])
    aguardar()
    pyautogui.write(row['Email'])
    aguardar()
    pyautogui.write(row['Celular'])
    aguardar()
    pyautogui.write(str(row['Data de Nascimento']), interval=0.2)
    aguardar()
    pyautogui.write(row['CPF'])
    aguardar()

    if   row['Atividade Extra'] == 'Dança':
        quant_tab = 0
    elif row['Atividade Extra'] == 'Muay Thai':
        quant_tab = 1
    elif row['Atividade Extra'] == 'Funcional':
        quant_tab = 2
    elif row['Atividade Extra'] == 'Boxe':
        quant_tab = 3
    elif row['Atividade Extra'] == 'Yoga':
        quant_tab = 4
    
    if quant_tab != 0:
        pyautogui.press("tab", presses = quant_tab)

    pyautogui.press("space")
    pyautogui.press("tab", presses = 5-quant_tab)
    wait()
    pyautogui.press("enter")
    wait()
```
