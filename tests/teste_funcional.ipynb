```python
import pyautogui, time, pandas as pd, pyperclip
arquivo = R'D:\Izadora\ADS\3 período - ADS\Teste-manutencao\Form-cadastro\invalidos-para-preencher.xlsx'
df = pd.read_excel(arquivo)
df['Data de Nascimento'] = df['Data de Nascimento'].dt.strftime('%d/%m/%Y')

def wait():
    time.sleep(0.3)

def repetir():
    wait()
    wait()
    pyautogui.click(1195,250)
    wait()
    pyautogui.press("tab", presses=1)
    wait()
    
def aguardar():
    pyautogui.press("tab")
    wait()

pyautogui.hotkey("ctrl","t")
wait()
pyautogui.write("http://127.0.0.1:5500")
wait()
pyautogui.press("enter")
wait()
wait()

for _, row in df.iterrows():
    repetir()
    pyautogui.write(row['Nome Completo'])
    aguardar()
    pyautogui.write(row['Email'])
    aguardar()
    pyautogui.write(row['Celular'])
    aguardar()
    pyautogui.write(str(row['Data de Nascimento']), interval=1)
    aguardar()
    pyautogui.write(row['CPF'])
    aguardar()

    if   row['Atividade Extra'] == 'Dança':
        quant_tab = 0
    elif row['Atividade Extra'] == 'Muay Thai':
        quant_tab = 1
    elif row['Atividade Extra'] == 'Funcional':
        quant_tab = 2
    elif row['Atividade Extra'] == 'Boxe':
        quant_tab = 3
    elif row['Atividade Extra'] == 'Yoga':
        quant_tab = 4
    
    if quant_tab != 0:
        pyautogui.press("tab", presses = quant_tab)

    pyautogui.press("space")
    pyautogui.press("tab", presses = 5-quant_tab)
    wait()
    pyautogui.press("enter", presses = 5)
    wait()

wait()
time.sleep(2)
pyautogui.click(400,580) #alterar conforme a tela
wait()
time.sleep(8)
pyautogui.hotkey('ctrl', 'l')
wait()
pyautogui.hotkey('ctrl', 'c')
wait()
endereço_copiado = pyperclip.paste()
wait()
endereço_esperado = "https://www.google.com/maps/place/Gym+Fit+academia/@-8.056332,-34.92451,15z/data=!4m6!3m5!1s0x7ab195a4a8d2a13:0xdaf1c0b5670fbd7f!8m2!3d-8.0563324!4d-34.9245096!16s%2Fg%2F11fn4zcgmg?hl=pt-BR&entry=ttu"

if endereço_copiado == endereço_esperado:
    print("Página redirecionada corretamente.")
else:
    print("Página não redirecionada corretamente.")
```
