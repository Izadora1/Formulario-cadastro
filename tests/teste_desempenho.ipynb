```python
import pyautogui, time, pandas as pd, pyperclip
arquivo = R'D:\Izadora\ADS\3 período - ADS\Teste-manutencao\Form-cadastro\dados-teste-desempenho.xlsx'
df = pd.read_excel(arquivo)
df['Aniversário'] = df['Aniversário'].dt.strftime('%d/%m/%Y')
cont = 1

def wait():
    time.sleep(0.3)

def repetir():
    global cont
    if cont == 1:
        pyautogui.click(557,158) #alterar conforme a tela
        wait()
        cont += 1
    else:
        pyautogui.click(1239,154) #alterar conforme a tela
        wait()
        cont -= 1
    pyautogui.press("tab", presses=1)
    wait()
    
def aguardar():
    pyautogui.press("tab")
    wait()

def abrirNavegador():
    pyautogui.press("win")
    wait()
    pyautogui.write("edge")
    time.sleep(0.8)
    pyautogui.press("enter")
    time.sleep(0.8)
    pyautogui.write("http://127.0.0.1:5500")
    wait()
    pyautogui.press("enter")
    time.sleep(0.5)

def preencher():
    for _, row in df.iterrows():
        repetir()
        pyautogui.write(row['Nome Completo'])
        aguardar()
        pyautogui.write(row['Email'])
        aguardar()
        pyautogui.write(row['Celular'])
        aguardar()
        pyautogui.write(str(row['Aniversário']), interval=0.7)
        aguardar()
        pyautogui.write(row['CPF'])
        aguardar()
        pyautogui.write(row['Rua'])
        aguardar()
        pyautogui.write(row['Bairro'])
        aguardar()
        pyautogui.write(row['Histórico Médico'])
        aguardar()
        pyautogui.write(row['Lesões'])
        aguardar()
        pyautogui.write(row['Objetivo'])
        aguardar()
        pyautogui.write(row['Treina?'])
        aguardar()
    
        if   row['Atividade Extra'] == 'Dança':
            quant_tab = 0
        elif row['Atividade Extra'] == 'Muay Thai':
            quant_tab = 1
        elif row['Atividade Extra'] == 'Funcional':
            quant_tab = 2
        elif row['Atividade Extra'] == 'Boxe':
            quant_tab = 3
        elif row['Atividade Extra'] == 'Yoga':
            quant_tab = 4
        
        if quant_tab != 0:
            pyautogui.press("tab", presses = quant_tab)
    
        pyautogui.press("space")
        pyautogui.press("tab", presses = 5-quant_tab)
        wait()
        pyautogui.press("enter", presses = 5)
        wait()

abrirNavegador()
wait()
abrirNavegador()
wait()
pyautogui.hotkey("alt", "tab")
wait()
pyautogui.hotkey("win", "left")
pyautogui.hotkey("win", "right")
wait()
pyautogui.press("enter")
time.sleep(0.8)
preencher()
```
